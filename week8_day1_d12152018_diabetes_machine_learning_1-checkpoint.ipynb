```python
diabetics = pd.read_csv("Diabetes_Data.csv", names=labels) 
```

```python
diabetics.head(2)
diabetics.columns
diabetics['Target'] = diabetics['Target'].astype("category")
diabetics['Target'].value_counts()
```

```python
numeric_vars = diabetics.drop('Target',axis=1)
```

```python
numeric_vars.head(2)
```

```python
# Divide data into Train and test
X_train, X_test = train_test_split(numeric_vars, test_size=0.150)
std= StandardScaler()
std.fit(X_train)
X_train_std = std.transform(X_train)
X_test_std = std.transform(X_test)
```

```python
# Number of clusters
kmeans = KMeans(n_clusters=2,random_state=1484641)
#Fiting the inputdata
kmeans = kmeans.fit(X_train_std)
# gettting the cluster labels
labels = kmeans.predict(X_train_std)
# centroid values
centroids = kmeans.cluster_centers_
```
