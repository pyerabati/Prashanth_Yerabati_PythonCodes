### Groupby and Aggregate
- groupby is one of the functions that can be applied on a data frame before aggregating(Remember the dplyr we group by a variable and then summarize on another numerical attribute. 
- We take a data, group by a categorical attribute and get aggregation on a numerical attribute

#### Example: We want to know how many matches are played in each city?
- We need to subset the records for each city and then count the number of records for each city i.e group by city get the records and count

```python
matches.groupby(by='city').aggregate(len)['id'] 
```

- Observe the syntax..
- Check the output if you have removed ['id'] from the above code.

```python
matches.groupby(by='city').aggregate(sum) #Just as a curius example changed len to sum and for all attributes
```

- You might have observed that the function sum on all the attributes is not meaningfull. See what happened to sum of id.
- Ideally we may want how many matches are played in a city, what is the average runs win or average wickets win, when is the most recent match played etc.. Which means that the aggregation for each variable is differnt. In such cases how do we work
- Given below is one such scenario

```python
matches.groupby('city').aggregate({
    'id': 'count',
    'win_by_runs': 'mean',
    'win_by_wickets': 'mean',
    'date': ['min', 'max']
})
```

#### What is the win percentage of `Chennai Super Kings`?

```python
csk = matches[(matches.team1 == 'Chennai Super Kings') | (matches.team2 == 'Chennai Super Kings')]['winner']
wins = sum(csk == 'Chennai Super Kings')
print('CSK has won {:.2f}% of matches'.format(100*wins/csk.shape[0]))
```
