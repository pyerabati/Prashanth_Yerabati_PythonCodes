```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, metrics
```

```python
# Load the boaston dataset
boston = datasets.load_boston(return_X_y=False)
```

```python
# Defining feature matrix(X) and response vector(y)
X = boston.data
y = boston.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting X and y into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
# Create a linear regression object
reg = linear_model.LinearRegression()
```

```python
# Train the model using train sets
reg_sco = reg.fit(X_train,y_train)
```

```python
# regression coefficients
print('Coefficients \n', reg.coef_)
```

```python
# Variance score: 1 means pefect predictions
print('variance score : {}'.format(reg.score(X_test,y_test)))
```

```python
# Setting plot style
plt.style.use('fivethirtyeight')
```

```python
# plotting residual errors on training data
plt.scatter(reg.predict(X_train),reg.predict(X_train) - y_train, color = "green", s=10, label = 'Train data')
```

```python
# plotting residual errors on test data
plt.scatter(reg.predict(X_test),reg.predict(X_test) - y_test, color = "blue", s=10, label = 'Test data')
```

```python
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
```

```python
## plotting legend 
plt.legend(loc = 'upper right') 
```

```python
## plot title 
plt.title("Residual errors") 
```

```python
## function to show plot 
plt.show() 
```
