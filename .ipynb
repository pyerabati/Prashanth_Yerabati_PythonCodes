```python
#### What is the win percentage of `Chennai Super Kings`?
```

```python
cs = matches[(matches.team1_id == 'CSK') | (matches.team2_id == 'CSK' ) ]
cswins = sum(cs.winner == 'CSK')
cswinsp = 100 * cswins/cs.shape[0]
print('CSK has won {:.2f}% of matches '.format(100*cswins/cs.shape[0]))
print(cswinsp)

```

```python
cs = matches[(matches.team1_id == 'CSK') | (matches.team2_id == 'CSK')]['winner']
wins = sum(cs == 'CSK')
print('CSK has won {:.2f}% of matches '.format(100*wins/cs.shape[0]))
```
